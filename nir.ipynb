```python
import pandas_datareader as pdr
from datetime import date

df = pdr.get_data_yahoo(symbols="TSLA", start=date(2021, 1, 4), end=date(2021, 10, 18))
```

```python
import plotly.graph_objects as go

go.Figure(data=go.Ohlc(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close']))
```

```python
df
```

```python
import pandas as pd
import numpy as np

diff = np.abs(df['High'] - df['Low'])

data = {'input':diff[:len(diff) - 1], 'output':diff[1:]}
data_train = pd.DataFrame(data=data)
```

```python
data_train['input']
```

```python
from sklearn.linear_model import LinearRegression


x = data_train["input"][:len(data_train) - 1].to_numpy().reshape((-1, 1))
y = data_train["output"][1:].to_numpy()
model = LinearRegression().fit(x, y)

r_sq = model.score(x, y)
info_reg = {
    'coeff_det': r_sq,
    'intercept' : model.intercept_,
    'slope' :  model.coef_ }

y_pred = model.predict(x)
y_pred = np.insert(y_pred, 0, np.NaN)
```

```python
y_pred
```

```python
data_train["lregg"] = pd.Series(list(y_pred), index=data_train.index)
data_train
```

```python
import plotly.express as px

x = data_train.index.to_numpy()
y = data_train['lregg'].to_numpy()

fig = px.scatter(data_train, x=data_train.index, y='output', opacity=0.65) # 
fig.add_traces(go.Scatter( x=x, y=y, name='Regression Fit'))
fig.show()
```

```python
output = data_train['output'].to_numpy()

fig = px.scatter(data_train, x='output', y='lregg', opacity=0.65)
fig.add_traces(go.Scatter( x=output, y=output))
fig.show()
```
